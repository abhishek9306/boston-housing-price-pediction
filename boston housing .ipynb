```python
# calculate the  correlation  between every coloumn on the data
corr = df.corr('pearson')

#take absolute values of correlations
corrs=[abs(corr[attr]['target']) for attr in list(features)]

#make a list of pairs [(corr ,features)]
l=list(zip(corrs,list(features)))

#sort the list of pairs in reverse /descending order 
#with the correlation valu as the key for sorting
l.sort(key=lambda x : x[0],reverse=True)

#unzip pairs to two lists
#zip(*l) - takes a list that looks like [[a,b,c],[d,e,f],[g,h,i]]
# and returns [[a,b,g],[b,e,f],[g,h,i]]
corrs, labels=list(zip((*l)))

#plot correlation with respect to the target  variables as a bargraph
index=np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variable')
plt.xticks(index,labels)
plt.show()
```

```python
#data preprocessing
# normalizing of data
x=df['LSTAT'].values
y=df['target'].values
```

```python
#before normalization
print(y[:5])
```

```python
x_scaler = MinMaxScaler()
x=x_scaler.fit_transform(x.reshape(-1,1))
x=x[:,-1]
y_scaler=MinMaxScaler()
y=y_scaler.fit_transform(y.reshape(-1,1))
y=y[:,-1]
```

```python
#after normalization
print(y[:5])
```

```python
#def error(m,x,c,t):
   # N = x.size
   # e = sum(((m*x+c)-t)**2)
   # return e * 1/(2*N)
    
```

```python
#spliting of the dat a set into training and testing dataset
#20% data is used for test and 80% is for training 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x+c)-t)*x)
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

```python
def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    m=init_m
    c=init_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print('error less than the threshold . Stoppimng gradient descent')
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
```

```python
%%time
init_m=0.9
init_c=0
learning_rate=0.001
iterations=250
error_threshold=0.001

m,c,error_values,mc_values=gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
#visuallizing the learning 
plt.scatter(xtrain,ytrain,color='b')
plt.plot(xtrain,(m*xtrain+c),color='r')
```

```python
#plot the error values
plt.plot(np.arange(len(error_values)),error_values)
plt.ylabel('error')
plt.xlabel('iterations')
```

```python
#step 5 prediction \
# after step 4 we have actual values of m,c so using these values we predicted the values in testing data
#we willthen evaluate the model with certain metrices  such as rmse,r2 score etc. the choice of these metrices will depend upon the type of data
predicted=(m*xtest)+c
```
